### Notebook to add admins to workspace
This notebook will replicate the permissions of whoever runs it, to all other email addresses on the list `user_list`. Be careful who you add, especially if running this as an admin.

You can also add users manually through the UI, by clicking on the top right Icon > `Settings` > Under `Workspace Admin` select `Users & Identity` > `Users` > `Add User`.

```python
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()

user_list = [
    "carlos.dip@databricks.com",
    "joao.gianfaldoni@databricks.com",
    "augusto.carneiro@databricks.com",
    "miqueiasamf@al.insper.edu.br",
    "aimeehanie@outlook.com",
]

needed_groups = w.current_user.me().groups.copy()
needed_entitlements = w.current_user.me().entitlements.copy()

for user in user_list:
    try:
        user_info =  w.users.create(
            user_name=user,
            entitlements=needed_entitlements,
            groups=needed_groups
        )
        print(f"User {user} not found, and created with admin permissions.")
    except Exception as e:
        if "already exists" in str(e).lower():
            print(f"User {user} found, skipping...")
        else:
            print(e)
```
